```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de las tiendas Alura Store
data = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingresos': [45000, 32000, 58000, 21000],
    'Evaluacion': [4.5, 3.8, 4.8, 3.1],
    'Costo_Envio': [150, 200, 120, 350],
    'Categoria': ['Electrónica', 'Ropa', 'Hogar', 'Juguetes']
}

df = pd.DataFrame(data)
print("¡Listo! Celda ejecutada con éxito.")
df
```

```python

import matplotlib.pyplot as plt


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# Gráfico 1: Ingresos (Facturación)
ax1.bar(df['Tienda'], df['Ingresos'], color='skyblue')
ax1.set_title('Facturación Total por Tienda')
ax1.set_ylabel('Ingresos ($)')

# Gráfico 2: Evaluaciones (Satisfacción)
ax2.bar(df['Tienda'], df['Evaluacion'], color='orange')
ax2.set_title('Promedio de Evaluación de Clientes')
ax2.set_ylabel('Rating (1-5)')

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.bar(df['Tienda'], df['Costo_Envio'], color='lightgreen')
plt.title('Costo Promedio de Envío por Tienda')
plt.ylabel('Costo ($)')
plt.show()
```
